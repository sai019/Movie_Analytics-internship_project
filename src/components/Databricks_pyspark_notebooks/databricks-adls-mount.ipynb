```
# Optionally, you can add <directory-name> to the source URI of your mount point.

container = dbutils.secrets.get(scope = 'project-scopes' , key = 'container')
storageaccount = dbutils.secrets.get(scope = 'project-scopes' , key = 'storageaccount')
accesskey = dbutils.secrets.get(scope = 'project-scopes' , key = 'accesskey')

dbutils.fs.mount(
  source= f"wasbs://{container}@{storageaccount}.blob.core.windows.net",
  mount_point= "/mnt/movie-analytics",
  extra_configs = {f"fs.azure.account.key.{storageaccount}.blob.core.windows.net":f"{accesskey}"})
```
